````python
def analyze_df(df: pd.DataFrame) -> None:
    # 1. Dimesionalidad del DataFrame
    print(f"Dimensionalidad del DataFrame: {df.shape}")

    # 2.Nombres de las columnas
    print(f"Nombres de las columnas: {df.columns.tolist()}")

    # 3.Muestra los primero 5, los ultimos 5 y 5 filas aleatorias del DataFrame
    print("Primeras 5 filas:")
    display(df.head())  # por defecto muestra las primeras 5 filas
    print("Últimas 5 filas:")
    display(df.tail())  # por defecto muestra las últimas 5 filas
    print("5 filas aleatorias:")
    display(df.sample(5))  # hay que especificar el número de filas aleatorias a mostrar

    # 4.Descripción rápida (variables númericas)
    print("\nDescripción estadística:")
    display(df.describe())

    # 5.1 Valores nulos por columna
    print(
        f"\nCantidad de valores nulos por columna:\n{df.isna().sum()}"
    )  # veo si es nula la cela y cuento la cantidad de valores nulos por columna
    # 5.2 Valores nulos en total
    print(f"Total valores nulos en el DataFrame: {df.isna().sum().sum()}")  # sumo sobre la suma para el total

    # 6. Valores únicos
    print(f"\nValores únicos por columna:\n{df.nunique()}")  # cuento la cantidad de valores únicos por columna

    # 7. Total de filas con nombre duplicados
    total_duplicados = df.duplicated(subset=["names"]).sum()
    print(f"\nTotal de filas con nombres duplicados: {total_duplicados}")

    # 8. Mostrar filas duplicadas
    if total_duplicados > 0:
        print("Filas duplicadas:")
        display(
            df[df.duplicated(subset=["names"], keep=False)]
        )  # keep=False cuento todas las filas que se duplican como duplicadas

    # 9.Histogramas
    px.histogram(df, x="math score", title="Distribución: Math Score").show()
    px.histogram(df, x="reading score", title="Distribución: Reading Score").show()
    px.histogram(df, x="writing score", title="Distribución: Writing Score").show()
````

**Respuesta:**

````python
def to_snake_case(s: str) -> str:
    # 1. Pasamos a minusculas
    # 2. Reemplazamos espacios por guiones bajos
    # 3. Quitamos la barra '/' si existe
    return s.lower().strip().replace(" ", "_").replace("/", "_")
````

````python
def clean_data(df: pd.DataFrame) -> pd.DataFrame:
    # Por seguridas hacemos una copia del DataFrame original
    df_clean = df.copy()

    # 1. Renombrar columnas a snake_case
    df_clean.columns = [to_snake_case(col) for col in df_clean.columns]

    # 2.Eliminar la palabra 'race' de la columna 'race/ethnicity' renombrandola a ethnicity
    df_clean = df_clean.rename(columns={"race_ethnicity": "ethnicity"})

    # 3. Forzar tipos numéricos y eliminar filas con datos no convertibles
    score_cols = ["math_score", "reading_score", "writing_score"]  # columnas de notas
    for col in score_cols:
        df_clean[col] = pd.to_numeric(
            df_clean[col], errors="coerce"
        )  # convierto a numérico, si no se puede convertir pongo NaN

    # 4. Eliminar filas con valores nulos (que no se pudieron convertir a numéricos)
    df_clean = df_clean.dropna()

    # 5. eliminar notas anómalas (menores a 0 o mayores a 100)
    for col in score_cols:
        df_clean = df_clean[(df_clean[col] >= 0) & (df_clean[col] <= 100)]  # solo dejos los datos con 0<= nota <=100

    # 6.Eliminar datos duplicados
    df_clean = df_clean.drop_duplicates(
        subset=["names"], keep="first"
    )  # elimino filas duplicadas basándome en la columna 'names' id único y mantengo la primera ocurrencia con keep='first'

    return df_clean
````

- Cabe destacar que es clave no seguir justamente el orden que se propone, ya que si se sigue así se comete el error de eliminar más registros de la cuenta, ya que si elimino por nombre duplicado primero y solo dejo uno de estos datos quizás el que dejo tiene sobre 100 puntos, oséa posteriormente lo eliminaría por completo del dataframe . Por lo que hay que eliminar duplicados de nombre al final del proceso solamente para asegurar de que lleguen todos los que tienes que estar

````python
df_cleaned = clean_data(df_grades)
df_cleaned
````
