### Extra: ¿Qué es el CI y qué verifican estas herramientas?

**CI (Integración Continua)** es una práctica en la que cada vez que se sube código al repositorio (mediante un push o un Pull Request), se ejecutan automáticamente una serie de verificaciones para asegurar que el código cumple ciertos estándares de calidad. En este curso, el CI está configurado en `.github/workflows/ci.yml` y es ejecutado por GitHub Actions.

Las herramientas que configuramos cumplen roles complementarios:

| Herramienta | ¿Qué verifica? |
|---|---|
| **`ruff` (linter)** | Detecta errores de estilo, imports no usados, variables indefinidas, patrones propensos a bugs y sintaxis desactualizada |
| **`ruff` (formatter)** | Formatea el código automáticamente para que tenga un estilo consistente (espacios, saltos de línea, comillas, etc.) |
| **`pre-commit`** | Ejecuta un conjunto de revisiones automáticamente *antes de cada commit*, impidiendo subir código que no pase las verificaciones |

En la práctica: cuando hagan `git commit`, `pre-commit` correrá `ruff` sobre sus archivos. Si hay problemas, el commit será bloqueado y deberán corregirlos antes de continuar. Esto asegura que todo el código que llegue al repositorio sea consistente y libre de errores básicos.

#### Pregunta 1 (0.5 pts)

Dada una lista de $n$ elementos, no necesariamente ordenada, defina funciones que busquen:
- El elemento menor.
- El elemento mayor.
- El promedio de los elementos en la lista.

Ejemplo:
```python
lista_n = [12,4,34,1,2,52,1,54,98,5,235,76,57,87,123,32,32]
menor = 1
mayor = 235
```


````python
# Al no poder usar funciones como max, min, sum, utilizamos los ciclos for para poder recorrer una lista


def elemento_menor(lista):
    menor = lista[0]  # Asumimos que el primeer elemento es el menor para empezar
    for n in lista:
        if n < menor:
            menor = n
    return menor


def elemento_mayor(lista):
    mayor = lista[0]
    for n in lista:
        if n > mayor:
            mayor = n
    return mayor


def promedio_elementos(lista):
    suma_total = 0
    cantidad_elementos = 0

    for n in lista:
        suma_total += n  # Vamos sumando cada número
        cantidad_elementos += 1  # Vamos contando la cantidad de elementos en la lista

        if cantidad_elementos == 0:  # Evitamos dividir por cero por si hay lista vacía
            return 0

    return suma_total / cantidad_elementos  # Vemos que debemos evitar los cero para que no falle
````

````python
# Pruebas
lista = [12, 4, 34, 1, 2, 52, 1, 54, 98, 5, 235, 76, 57, 87, 123, 32, 32]

print(elemento_menor(lista))
print(elemento_mayor(lista))
print(promedio_elementos(lista))
````

#### Pregunta 2 (0.5 pts)
Genere una función que obtenga la moda de una lista usando diccionarios.

````python
# Nuevamente usamos ciclos for para ver el elemento que más se repite en un diccionario
def obtener_moda(lista):
    if len(lista) == 0:  # Si el largo de la lista es cero no hay moda
        return None

    diccionario_conteos = {}  # Creamos diccionario vacío de conteo

    # Contar cuántas veces aparece cada elemento
    for elemento in lista:
        if elemento in diccionario_conteos:
            diccionario_conteos[elemento] += 1  # Ya existía el elemento en el diccionario le sumamos 1
        else:
            diccionario_conteos[elemento] = 1  # Si el elemento en el diccionario es nuevo empezamos en 1

    # Buscar quién tiene el conteo más alto en el diccionario
    moda = None
    max_apariciones = 0

    for (
        numero,
        cantidad,
    ) in diccionario_conteos.items():  # Buscamos en los items del diccionario el ganador lee {número, cantidad}
        if cantidad > max_apariciones:
            max_apariciones = cantidad
            moda = numero

    return moda
````

````python
# Pruebas
lista_2 = [1, 1, 2, 2, 3, 3, 3, 4, 5]

print(obtener_moda(lista_2))
````

#### Pregunta 3 (0.5 pts)
Vea si un número es palíndromo o no. Los números palíndromos son aquellos que al leerlos al revés y al derecho son iguales, por ejemplo el 123321 o el 98789.

````python
# Podemos trabajar con índices y comparlos entre dos palabras para ver si son palíndromos


def es_palindromo(numero):
    texto = str(numero)  # Transformamos a string para usar indices

    # Inicializamos los índices en los extremos
    indice_izq = 0
    indice_der = len(texto) - 1

    # Iteramos acercando los índices hacia el centro
    while indice_izq < indice_der:
        if texto[indice_izq] != texto[indice_der]:  # Falla la condición de palíndromo
            return False
        # Acercamos hacia el centro
        indice_izq += 1
        indice_der -= 1

    return True
````

````python
# Prueba
print(es_palindromo(445566665544))  # Palíndromo
print(es_palindromo(22332123453))  # No Palíndromo
````

#### Pregunta 4 (0.5 pts)
Dado un texto, conviértalo en estilo Camel Case, vea el siguiente ejemplo:

    Texto inicial: “Este es un texto de ejemplo para mostrar el Camel Case”.

    Texto final: “EsteEsUnTextoDeEjemploParaMostrarElCamelCase”.

````python
# Podemos separar el string con la función split() y unir


def texto_a_camel_case(texto):
    # split() para separar la frase por los espacios y nos entrega una lista de palabras.
    lista_de_palabras = texto.split()

    # String vacío donde iremos pegando las palabras finales
    resultado_final = ""

    # Limpiar y Unir
    for palabra in lista_de_palabras:
        # capitalize() deja la 1ra letra en mayúscula y fuerza el resto a minúscula.
        palabra_lista = palabra.capitalize()

        # Pegamos la palabra ya formateada a nuestro resultado final (sin espacios)
        resultado_final += palabra_lista

    return resultado_final
````

````python
# Prueba
frase_1 = "Esto es una prueba para ver que tal funciona"

texto_a_camel_case(frase_1)
````

#### Pregunta 1 (0.5 pts):

Un monto inicial de dinero (por ejemplo un millón de pesos) se puede depositar en un banco con un interés anual de 5%. Esto significa  que  cumplido  un  año  se  dispondrá  de  \\$1.050.000. Si se deposita por dos años, entonces  al  finalizar  el  2º  año se dispondrá de \\$1.102.500,  es  decir,  \\$1.050.000  reajustado  en  5%.  

Al  respecto, si  se  invoca  a  la  función  `calc_capital_final(1000000, 5, 2)`  entrega  el resultado `1050000.0`. 

Escriba la función `calc_capital_final`

$$C_f = C_i(1+i)^n$$

- $C_f$ = Capital Final
- $C_i$ = Capital Inicial
- $i$ = Tasa de interés
- $n$ = Periodo del ahorro

Además, incluya tests que certifiquen el correcto funcionamiento de la función.

````python
def calc_capital_final(capital_inicial, tasa_interes, años):
    i = tasa_interes / 100  # Convertimos la tasa de porcentaje a decimal

    capital_final = capital_inicial * (1 + i) ** años  # Fórmula del interés compuesto
    return capital_final
````

#### Pregunta 2 (0.5 pts):

Un anagrama es una palabra que resulta de la transposición de letras de otra palabra o frase.

Por ejemplo, las palabras “caso” y “caos” sonan agramas de “saco”  y las palabras “acaso” y “ocaso” no son anagramas de “caso”. Las anagramas poseen las mismas letras, con la misma cantidad de apariciones, pero en un orden diferente

Escriba la función `def anagrama(x,y)` que reciba dos palabras y entregue `True` si son anagramas y `False` si no.


````python
def anagrama(x, y):
    # Normalizar strings por si hay mayúsculas
    str1 = x.lower()
    str2 = y.lower()

    # Si no miden lo mismo no son anagramas
    if len(str1) != len(str2):
        return False

    # Ordenar alfabetcicamente los caracteres de cada string
    sorted_str1 = sorted(str1)
    sorted_str2 = sorted(str2)

    # Comparar los strings ordenados
    return sorted_str1 == sorted_str2
````

````python
# Pruebas
print(anagrama("caso", "Caos"))  # True
print(anagrama("acaso", "ocaso"))  # False
````

Primero, se identificó que el error proviene del archivo `animales.py`. El problema lógico radica en que se utilizan múltiples sentencias `if` consecutivas en lugar de `elif`. Esto provoca que el programa evalúe todas las condiciones hacia abajo y, a menos que el usuario ingrese exactamente la última opción válida ('fox'), el código siempre terminará cayendo en el bloque `else`, levantando el error `ValueError('animal no reconocido :(')`.

Para encontrar al responsablede este error, se ejecutó el comando `git blame animales.py`, el cual inspecciona la autoría del código línea por línea. La terminal revela de forma clara que estas líneas defectuosas fueron introducidas por el autor **mezosky**.

5. Realice una `merge` de la rama `fix-ifs` con la rama `main` y describa lo que ocurre. ¿Qué parte del código podría estar ocasionando esto? (0.25 pts)

Al intentar fusionar la rama `fix-ifs` hacia la rama `main`, la terminal arrojó un mensaje de error indicando un conflicto de contenido: `CONFLICT (content): Merge conflict in main.py`.

Este conflicto se produce por lo que se podía suponer porque Git detecta que ambas ramas modificaron exactamente la misma porción de código en la historia, pero con acciones contradictorias:
1. En la rama `main` (donde estamos parados), la lógica de los animales fue eliminada de `main.py` para ser modularizada en un nuevo archivo llamado `animales.py`.
2. En la rama remota `fix-ifs`, esa misma lógica fue editada (se corrigieron los `if` por `elif`) directamente dentro del archivo original `main.py`.

Como Git no sabe qué hacer ni tiene la capacidad de decidir por sí solo si debe quedarse con la nueva estructura de módulos o con la corrección de los `ifs`, simplemente se confunde. Por eso, frena la fusión automática (Auto-merging) y deja el proceso en pausa para que nosotros resolvamos el conflicto a mano.
